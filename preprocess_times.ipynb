```python
import dask.dataframe as dd
import pandas as pd

# 分块读入CSV文件
df_a = dd.read_csv('/Users/rainfow/Desktop/modified_data/combined_nspl.csv', blocksize=500000)
df_b = dd.read_csv('/Users/rainfow/Desktop/Summer projects/companies_data/BasicCompanyData-part7_7.csv', blocksize=500000) 

print(len(df_a))
print(len(df_b))

df_b['lat'] = None
df_b['long'] = None

# 设置索引  
df_a = df_a.set_index('pcds')
df_b = df_b.set_index('RegAddress.PostCode')

print(len(df_a))
print(len(df_b))

# 提取需要的Series
lat_series = df_a['lat']
long_series = df_a['long']

# 定义映射函数
def map_func(df, lat_series, long_series):
    df['lat'] = df.index.map(lat_series.to_dict()).fillna(None)
    df['long'] = df.index.map(long_series.to_dict()).fillna(None)
    return df

# 并行应用映射函数
df_b = df_b.map_partitions(map_func, lat_series=lat_series, long_series=long_series, meta=df_b)
print(len(df_b))
print(df_b)

# 触发执行
df_b = df_b.compute()

# 输出CSV文件
df_b.to_csv('/Users/rainfow/Desktop/Summer projects/companies_data/merged_7_7.csv', index=False)
```

```python
import pandas as pd

# 读取两个CSV文件
df_a = pd.read_csv('/Users/rainfow/Desktop/modified_data/combined_nspl.csv')
df_b = pd.read_csv('/Users/rainfow/Desktop/Summer projects/companies_data/postcode-part1_7.csv')

# 检查 df_a 的结构
# print(df_a.head())  # 这将显示 DataFrame 的前几行数据
# print(df_a.columns)  # 这将显示列名

# 检查 df_b 的结构和列名
print(df_b.head())  # 这将显示 DataFrame 的前几行数据
print(df_b.columns)  # 这将显示列名

# 检查 df_a 和 df_b 的数据类型
print(df_a.dtypes)
print(df_b.dtypes)

# 检查 df_a 中是否存在 'pcds' 列
if 'pcds' not in df_a.columns:
    raise ValueError("在 df_a 中找不到 'pcds' 列。")

# 构建邮编与坐标映射的字典
postcode_map = df_a.set_index('pcds')[['lat', 'long']].to_dict('index')

# 使用 map 方法结合 lambda 函数进行映射
df_b['lat'] = df_b['RegAddress.PostCode'].map(lambda x: postcode_map.get(x, {}).get('lat', None))
df_b['long'] = df_b['RegAddress.PostCode'].map(lambda x: postcode_map.get(x, {}).get('long', None))

# 输出结果到文件
df_b.to_csv('/Users/rainfow/Desktop/Summer projects/companies_data/merged_1_7.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 假设found_year_counts是之前计算得到的公司成立年份的数量分布
# 假设该数据列名为"found_year"和"companies_number"，为时间序列数据

ones_tens_hundreds = (found_year_counts["companies_number"] >= 0) & (found_year_counts["companies_number"] < 1000)
thousands = (found_year_counts["companies_number"] >= 1000) & (found_year_counts["companies_number"] < 10000)
ten_thousands = (found_year_counts["companies_number"] >= 10000) & (found_year_counts["companies_number"] < 100000)
hundred_thousands = (found_year_counts["companies_number"] >= 100000) & (found_year_counts["companies_number"] < 1000000)

# 绘制图表
fig, ax = plt.subplots(figsize=(12, 8))

# 绘制折线图并添加label进行说明
ax.set_title("Number of Companies Founded")
ax.plot(found_year_counts[ones_tens_hundreds]["found_year"], found_year_counts[ones_tens_hundreds]["companies_number"], label="ones_tens_hundreds", marker='o')
ax.plot(found_year_counts[thousands]["found_year"], found_year_counts[thousands]["companies_number"], label="thousands", marker='o')
ax.plot(found_year_counts[ten_thousands]["found_year"], found_year_counts[ten_thousands]["companies_number"], label="ten_thousands", marker='o')
ax.plot(found_year_counts[hundred_thousands]["found_year"], found_year_counts[hundred_thousands]["companies_number"], label="hundred_thousands", marker='o')

# 设置y轴刻度范围
ax.set_ylim(0, 800000)

# 设置x轴和y轴标签
ax.set_xlabel("Year")
ax.set_ylabel("Number of Companies")

# 添加图例
ax.legend()

# 显示网格线
ax.grid(True)

# 寻找显著变化趋势的点
company_counts = found_year_counts["companies_number"]
second_derivative = np.gradient(np.gradient(company_counts))  # 计算二阶导数

# 设置阈值，用于筛选显著变化的点
threshold = 10000

# 找到显著变化趋势的点的索引
significant_points = np.where(np.abs(second_derivative) > threshold)[0]

# 在图上标记显著变化趋势的点，并显示具体年份信息
for index in significant_points:
    year = found_year_counts.iloc[index]["found_year"]
    count = company_counts.iloc[index]
    ax.annotate(f"Year: {year}\nCount: {count}", xy=(year, count),
                xytext=(-30, 15), textcoords='offset points', arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))

# 显示图表
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft
from statsmodels.graphics.tsaplots import plot_acf

# 进行傅里叶变换
companies_number_fft = fft(found_year_counts["companies_number"])
power_spectrum = np.abs(companies_number_fft) ** 2
frequencies = np.fft.fftfreq(len(found_year_counts), 1)  # 获取对应的频率值

# 绘制傅里叶变换的结果（频谱图）
plt.figure(figsize=(10, 6))
plt.plot(frequencies, power_spectrum)
plt.xlabel("Frequency")
plt.ylabel("Power Spectrum")
plt.title("Fourier Transform - Power Spectrum")
plt.grid(True)
plt.show()

# 使用自相关函数检测数据的周期性
plot_acf(found_year_counts["companies_number"], lags=30)
plt.xlabel("Lag")
plt.ylabel("Autocorrelation")
plt.title("Autocorrelation Function")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 假设found_year_counts是之前计算得到的公司成立年份的数量分布
# 假设该数据列名为"found_year"和"Number_of_Companies"，为时间序列数据

# 创建时间序列数据，只选择1980年以前的数据
data_1945_before = found_year_counts[found_year_counts["found_year"] <= 1945]["companies_number"]

# 创建时间序列数据，选择1980-1999年的数据
data_1945_1985 = found_year_counts[(found_year_counts["found_year"] > 1945) & (found_year_counts["found_year"] <= 1985)]["companies_number"]

# 创建时间序列数据，选择2000-2011年的数据
data_1985_2010 = found_year_counts[(found_year_counts["found_year"] > 1985) & (found_year_counts["found_year"] <= 2010)]["companies_number"]

# 创建时间序列数据，选择2012年以后的数据
data_2010_after = found_year_counts[found_year_counts["found_year"] > 2010]["companies_number"]

# 分别进行傅里叶变换并绘制频谱图
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

for ax, data, title in zip(axs.flat, [data_1945_before, data_1945_1985, data_1985_2010, data_2010_after],
                           ["1945 and Before", "1945-1985", "1985-2010", "2010 and After"]):
    # 进行傅里叶变换
    fft_result = np.fft.fft(data)

    # 计算频率
    n = len(data)
    freq = np.fft.fftfreq(n)

    # 绘制频谱图
    ax.plot(freq, np.abs(fft_result))
    ax.set_xlabel("Frequency")
    ax.set_ylabel("Amplitude")
    ax.set_title(f"Frequency Spectrum ({title})")
    ax.grid()

# 自动调整子图间的间隔，防止重叠
plt.tight_layout()

# 显示图表
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# 假设found_year_counts是之前计算得到的公司成立年份的数量分布
# 假设该数据列名为"found_year"和"Number_of_Companies"，为时间序列数据


data_1945_before = found_year_counts[found_year_counts["found_year"] <= 1945]["companies_number"]
data_1945_1985 = found_year_counts[(found_year_counts["found_year"] > 1945) & (found_year_counts["found_year"] <= 1985)]["companies_number"]
data_1985_2010 = found_year_counts[(found_year_counts["found_year"] > 1985) & (found_year_counts["found_year"] <= 2010)]["companies_number"]
data_2010_after = found_year_counts[found_year_counts["found_year"] > 2010]["companies_number"]
# Create subplots for four ACF graphs
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot ACF for each time period
for ax, data, title in zip(axs.flat, [data_1945_before, data_1945_1985, data_1985_2010, data_2010_after],
                           ["data_1945_before", "data_1945_1985", "data_1985_2010", "data_2010_after"]):
    
    # Calculate the maximum number of lags based on the data length
    max_lags = len(data) - 1

    # Plot ACF
    sm.graphics.tsa.plot_acf(data, lags=max_lags, ax=ax)
    ax.set_xlabel("Lag")
    ax.set_ylabel("Autocorrelation")
    ax.set_title(f"Autocorrelation Function (ACF) - {title}")
    ax.grid()

# Automatically adjust the layout to prevent overlapping
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# 使用孤立森林进行时间序列异常检测
from sklearn.ensemble import IsolationForest

# 假设found_year_counts是之前计算得到的公司成立年份的数量分布
# 假设该数据列名为"Found_Year"和"Number_of_Companies"，为时间序列数据

# 创建X矩阵，其中只包含"Number_of_Companies"列的值
X = found_year_counts["companies_number"].values.reshape(-1, 1)

# 使用孤立森林算法进行异常检测
clf = IsolationForest(contamination=0.05, random_state=42)
found_year_counts["Is_Outlier"] = clf.fit_predict(X)

print(found_year_counts)

print(found_year_counts.loc[found_year_counts["Is_Outlier"] == -1])
```

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller


data = found_year_counts["companies_number"]
# 假设data是时间序列数据，可以是一个Series或DataFrame的一列
result = adfuller(data)

# 提取ADF检验的结果
adf_statistic = result[0]
p_value = result[1]
critical_values = result[4]

# 输出ADF检验的结果
print("ADF Statistic:", adf_statistic)
print("p-value:", p_value)
print("Critical Values:", critical_values)
```
